```python
from pylab import *
from scipy.integrate import odeint
%matplotlib inline
```

```python
# define differential equations
def eqn1(x,t):
    x_dot = 3*exp(-t)
    return x_dot

def eqn2(y,t):
    y_dot = 3 - y
    return y_dot
```

```python
tmax = [1,2,4,8]
for i in range(len(tmax)):
    # define initial values
    x0 = 0
    # define time t
    t = arange(0,tmax[i],0.1)
    # solve the differential equation
    x = odeint(eqn1,x0,t)
    subplot(2,2,i+1)
    # plot the output
    plot(t,x)
    grid()
    xlabel('time')
    ylabel('x')
    
```

```python

for i in range(len(tmax)):
    # define initial values
    y0 = 0
    # define time t
    t = arange(0,tmax[i],0.1)
    # solve the differential equation
    y = odeint(eqn2,y0,t)
    subplot(2,2,i+1)
    # plot the output
    plot(t,y)
    grid()
    xlabel('time')
    ylabel('y')
```

# From the two set of plots, we see that both solutions are equivalent and both x and y converges to 3 while t goes to infinity.
