```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Construct lines
x = np.linspace(start=0, stop=20, num=1000)
y1 = (25.0 - x)/2.0
y2 = (2*x - 8.0)/4.0
y3 = 2*x - 5.0
y4 = 2.0*np.ones(len(x))
```

```python
# 2. Make plot
plt.plot(x,y1,label='Line-1')
plt.plot(x,y2,label='Line-2')
plt.plot(x,y3,label='Line-3')
plt.plot(x,y4,label='Line-4')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.xlim([0,20])
plt.ylim([0,20])

# 3. Fill feasible region
y5 = np.minimum(y1,y3)
y6 = np.maximum(y2,y4)
xlimit = np.logical_and(x>=3.5, x<=14.5)
plt.fill_between(x,y5,y6,where=xlimit, color='grey')
plt.show()
```
